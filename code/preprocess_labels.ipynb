```python
def findnth(haystack, needle, n):
    """Find Nth occurance of string in another string"""
    parts= haystack.split(needle, n+1)
    if len(parts)<=n+1:
        return -1
    return len(haystack)-len(parts[-1])-len(needle)

def calc_frames(barcodes, include_dx=True):
    import os
    data_folder = '/var/shared/hist_project/Data/Level2/TCGA'
    image_names = os.listdir(data_folder)
    frame_dict = FrameCountCache([ImageReader.open(data_folder+'/'+img_name,type='png') for img_name in image_names]).dict
    sum_dict = {b:0 for b in barcodes}
    for img_name,frames in frame_dict.iteritems():
        if not include_dx and '-DX' in img_name:
            continue
        tcga_barcode = img_name[:findnth(img_name,'-',2)]
        sum_dict[tcga_barcode] += frames
    
    return [sum_dict[b] for b in barcodes]

def calc_slides(barcodes, include_dx=True):
    import os
    data_folder = '/var/shared/hist_project/Data/Level2/TCGA'
    image_names = os.listdir(data_folder)
    sum_dict = {b:0 for b in barcodes}
    for img_name in image_names:
        if not include_dx and '-DX' in img_name:
            continue
        tcga_barcode = img_name[:findnth(img_name,'-',2)]
        sum_dict[tcga_barcode] += 1
    
    return [sum_dict[b] for b in barcodes]

def unify_stage(stages):
    import re
    
    rep = {"ia": "i", "ib": "i", "ic":"i", "va":"v", "vb":"v", "vc":"v",
          "i/ii nos":"UNK", "not reported": "UNK", "stage 0":"UNK", "stage x":"UNK"} # define desired replacements here
    # use these three lines to do the replacement
    rep = dict((re.escape(k), v) for k, v in rep.iteritems())
    pattern = re.compile("|".join(rep.keys()))
    
    return [pattern.sub(lambda m: rep[re.escape(m.group(0))], stage) for stage in stages]

include_dx=False
metadata_all_2 = metadata_all.assign(Behaviour=pd.Series(list(metadata_all.Morphology.str[-1:])).values,
                                     CellType=pd.Series(list(metadata_all.Morphology.str[:4])).values,
                                     Tumor_Stage_U=pd.Series(unify_stage(list(metadata_all.Tumor_Stage.fillna('not reported')))).values,
                                     Frames=pd.Series(calc_frames(list(metadata_all.Barcode),include_dx=include_dx)).values,
                                     Slides=pd.Series(calc_slides(list(metadata_all.Barcode),include_dx=include_dx)).values)
metadata_all_2 = metadata_all_2.filter(['Barcode','PrimarySite','Tumor_Stage_U','CellType','Behaviour','Tissue_Organ_Of_Origin','Cancer_type', 'Frames','Slides'])
metadata_all_2 = metadata_all_2.loc[metadata_all_2['Frames'] > 0]

print len(metadata_all_2)

display(metadata_all_2.head(n=10))
print '\nWe only use: Barcode, PrimarySite, Tumor_Stage_U, Frames'
```

```python
def diff_versions(df_new, df_old):
    
    new_keys_x = set(df_new['Barcode'])
    
    prob_cols = ['ER_Status','PR_Status','HER2_Final_Status'][2:3]
    
    df_new = df_new.filter(['Barcode']+prob_cols)
    df_old = df_old.filter(['Barcode']+prob_cols)
    
    # Standarize
    for col in prob_cols:
        df_old.loc[df_old[col].isin(['Not_Performed', 'Performed but Not Available', 'Not Performed']), col] = '[Not Available]'
    df_old = df_old.fillna('[Not Available]')
    
    df_new = df_new.query(' or '.join(['%s in ("Positive","Negative","Equivocal","Indeterminate")' % col for col in prob_cols]))
    df_old = df_old.query(' or '.join(['%s in ("Positive","Negative","Equivocal","Indeterminate")' % col for col in prob_cols]))
    
    old_len = len(df_old)
    new_len = len(df_new)
    
    old_keys = set(df_old['Barcode'])
    new_keys = set(df_new['Barcode'])
    added_keys = new_keys - old_keys
    deleted_keys = old_keys - new_keys
    
    added = df_new[df_new.Barcode.isin(added_keys)]
    deleted = df_old[df_old.Barcode.isin(deleted_keys)]
    added_len = len(added)
    deleted_len = len(deleted)
    
    joined = pd.merge(df_new, df_old, on=["Barcode"],how='inner',suffixes=['_new','_old'])
    joined_len = len(joined)
    
    
    same = joined
    for col in prob_cols:
        same = same.loc[same[col+'_new']==same[col+'_old']]
    different = joined.query(' or '.join(['%s != %s' % (col+'_new',col+'_old') for col in prob_cols]))
    
    same_len = len(same)
    different_len = len(different)
    
    print 'old(%d), new(%d), joined(%d), same(%d), diff(%d), added(%d), deleted(%d)' % (old_len,new_len,joined_len, same_len, different_len, added_len, deleted_len)
        
    return different, added, deleted

different, added, deleted = diff_versions(breastr_genomic_meta_2018, breastr_genomic_meta)

print '\nadded'
display(added.head())

print '\ndeleted'
display(deleted.head())

print '\ndifferent'
display(different.head())
```

```python
import numpy as np
import random

class Patient(object):
    def __init__(self, barcode, domains):
        self.barcode = barcode
        self.domains = domains
    def len_domains(self):
        return sum([1 for v in self.domains.values() if v != 'UNK'])
    def can_add(self, counts, limits, order=1, patience=0):
        need_me = False
        for k,v in self.domains.iteritems():
            if v == 'UNK':
                continue
            if v not in counts[k]:
                counts[k][v] = 0
            if limits[k] > counts[k][v]:
                need_me = True
            if min(limits[k] * (1.0 + patience) ,order) <= counts[k][v]:
                return False
        if not need_me:
            return False
        for k,v in self.domains.iteritems():
            if v != 'UNK':
                counts[k][v] += 1
        return True
    def __str__( self ):
        return self.barcode + ' , ' + str(self.domains)
    def __repr__( self ):
        return str(self)

def split_train_test(meta_join):
    test_patients = []
    patients = []
    max_domains_num = 7
    for barcode, group in meta_join.groupby(['Barcode']):
        has_normal = 'UNK'
        #has_diff = 'UNK'
        has_er = 'UNK'
        has_her2_fish = 'UNK'
        has_her2_ihc = 'UNK'
        has_her2_ihc_score = 'UNK'
        has_pr = 'UNK'
        site = None
        for i, row in group.iterrows():
            site = site or row.PrimarySite
            cancer_type = row.Cancer_type
            # order is important
            if cancer_type in sample_cancers:
                if row.Sample_type == 'Normal':
                    has_normal = (cancer_type,'Normal')
            if site in er_sites:
                #has_diff = has_diff if has_diff != 'UNK' else row.Differentiated
                has_er = has_er if has_er != 'UNK' else row.ER_Status
                has_her2_fish = has_her2_fish if has_her2_fish != 'UNK' else row.HER2_Fish
                has_her2_ihc = has_her2_ihc if has_her2_ihc != 'UNK' else row.HER2_IHC
                has_her2_ihc_score = has_her2_ihc_score if has_her2_ihc_score != 'UNK' else row.HER2_IHC_SCORE_FINAL
                has_pr = has_pr if has_pr != 'UNK' else row.PR_Status
            if cancer_type not in cancer_types:
                cancer_type = 'UNK'
            
        domains = {'PrimarySite':site,
                   'Sample_type':has_normal,
                   'HER2_Fish':has_her2_fish,
                   #'HER2_IHC':has_her2_ihc,
                   'HER2_IHC_SCORE_FINAL':has_her2_ihc_score,
                   'PR':has_pr,
                   'ER':has_er,
                   'Cancer_type' : cancer_type
                   }
        #domains = {k+'_'+v for k,v in domains.iteritems() if v!='UNK'}
        patients.append(Patient(barcode, domains))
    
    counts = {'PrimarySite':{},
              'Sample_type':{},
              'HER2_Fish':{},
              #'HER2_IHC':{},
              'HER2_IHC_SCORE_FINAL':{},
              'PR':{},
              'ER':{},
              'Cancer_type' : {}
              }
    limits = {'PrimarySite':80,
              'Sample_type':30,
              'HER2_Fish':25,
              #'HER2_IHC':30,
              'HER2_IHC_SCORE_FINAL':40,
              'PR':40,
              'ER':40,
              'Cancer_type' : 30
              }
    #random.shuffle(patients)
    test_barcodes = set()
    priorities = [lambda p: p.domains['HER2_Fish']=='Positive',
                 #lambda p : p.domains['HER2_IHC']=='Negative',
                 lambda p : p.domains['HER2_Fish']=='Negative']
    for i in range(max_domains_num,0,-1):
        filtered_patients = [p for p in patients if p.len_domains()==i and p.barcode not in test_barcodes]
        filtered_patients_priorities = [[p for p in filtered_patients if pr(p)] for pr in priorities]
        filtered_patients_priorities.append([p for p in filtered_patients if not any([pr(p) for pr in priorities])])

        print len(filtered_patients)
            
        for sub_patients in filtered_patients_priorities:
            for p in sub_patients:
                if p.can_add(counts, limits,order=100,patience=.2):
                    test_patients.append(p)
                    test_barcodes.add(p.barcode)
    
    print counts
    return test_patients
    
test_patients = split_train_test(meta_join_org)

#print patients_per_site

test_cases = set().union([p.barcode for p in test_patients])

print '%d cases' % len(test_cases)
print 'Done'
```

```python
import scipy.io

def create_matlab_equations(slide_df, frame_df, min_g_size, n_patches):
    def get_var_index(key ,var_map, verbose=False):
        if key in var_map:
            next_id = var_map[key]
        else:
            next_id = len(var_map)
            var_map[key] = next_id
        return next_id, var_map

    def hash_dict(my_dict):
        return frozenset(my_dict.items())

    def build_equation(prob_col, prob_sub_col, problem_sub_cols, dataframe):
        import itertools

        other_cols = problem_sub_cols.keys()
        other_cols.remove(prob_col)

        other_sub_cols = {col: (list(problem_sub_cols[col]) + ['UNK']) for col in other_cols}

        equation_dict = {}
        for element in itertools.product(*[other_sub_cols[col] for col in other_cols]):
            triplet_dict = {col:e  for col, e in zip(other_cols, element)}
            triplet_dict.update({prob_col : prob_sub_col})
            #print triplet_dict
            triplet_group_size = get_sub_group_size(triplet_dict, dataframe)
            #print triplet_group_size
            if triplet_group_size != 0:
                equation_dict[hash_dict(triplet_dict)] = triplet_group_size
        equation_dict[prob_col] = -1
        return equation_dict

    def named_eq_to_id_eq(eq, var_map, verbose=False):
        eq_id_dict = {}
        for key,coeff in eq.iteritems():
            var_id, var_map = get_var_index(key, var_map, verbose)
            eq_id_dict[var_id] = coeff
        return eq_id_dict, var_map

    #print build_equation('Sample_type', 'Tumor', problem_sub_cols, count2)

    equations = []
    for prob_col in problem_cols: # PrimarySite, SampleType, Differentiated
        for prob_sub_col in problem_sub_cols[prob_col]: # Normal \ Tumor
            eq = build_equation(prob_col, prob_sub_col, problem_sub_cols, slide_df)
            equations.append(eq)

    #print equations[0]

    equations_ids = []
    var_map = {}
    for eq in equations:
        eq_id, var_map = named_eq_to_id_eq(eq, var_map, verbose=True)
        equations_ids.append(eq_id)

    var_map_inv = {indx:key for key,indx in var_map.iteritems()}

    n_equations = len(equations_ids)
    n_vars = len(var_map)
    C = np.zeros(shape=(n_equations, n_vars), dtype=float)
    d = np.zeros(n_equations, dtype=float)
    lb = np.ones(n_vars, dtype=float) * 10
    ub = np.array([1.0 * get_sub_group_size(dict(var_map_inv[i]), frame_df) / get_sub_group_size(dict(var_map_inv[i]), slide_df) if type(var_map_inv[i]) == frozenset else float('Inf') for i in range(n_vars)])  #* 500 #float('Inf')
    #ub = np.ones(n_vars, dtype=float) * 900 #float('Inf')
    
    for i, eq_id in enumerate(equations_ids):
        for var_id, coeff in eq_id.iteritems():
            C[i,var_id] = coeff

    for i in range(n_vars):
        if type(var_map_inv[i]) != frozenset: # problem quantity var
            num_sub_cols = len(problem_sub_cols[var_map_inv[i]])
            lb[i] = min_g_size #/ num_sub_cols # try and error to set it var. highest with possible solution
    
    # Make solution divisible by 2 (n_patches)
    lb = (lb.astype(int) / n_patches).astype(float) # !!!
    ub = (ub.astype(int) / n_patches).astype(float) # !!!
    C = (C.astype(int) / n_patches).astype(float) # !!!
    d = (d.astype(int) / n_patches).astype(float) # !!!
    for i in range(n_vars):
        if type(var_map_inv[i]) != frozenset: # problem quantity var
            ub[i] = float('Inf')

    matlab_dict = { 'C': C, 'lb':lb, 'ub':ub, 'd':d }
    return matlab_dict, var_map, var_map_inv
    
data_matlab_dict, data_var_map, data_var_map_inv = create_matlab_equations(data_slide_df, data_frame_df, 20000, 3) # ALL
train_matlab_dict, train_var_map, train_var_map_inv = create_matlab_equations(train_slide_df, train_frame_df, 15000, 3) # Train
test_matlab_dict, test_var_map, test_var_map_inv = create_matlab_equations(test_slide_df, test_frame_df, 3000, 3) # Test

#print train_var_map
print train_matlab_dict
scipy.io.savemat('/var/shared/hist_project/Data/tmp/data_matlab_3patches_er_btumor.mat', mdict=data_matlab_dict)
scipy.io.savemat('/var/shared/hist_project/Data/tmp/train_matlab_3patches_er_btumor.mat', mdict=train_matlab_dict)
scipy.io.savemat('/var/shared/hist_project/Data/tmp/test_matlab_3patches_er_btumor.mat', mdict=test_matlab_dict)
```
